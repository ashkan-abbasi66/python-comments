```python
import torch
from torch.autograd import Variable

dtype = torch.FloatTensor

# x = Variable(torch.from_numpy(Xtr).type(dtype), requires_grad=False)
# y = Variable(torch.from_numpy(ytr).type(dtype), requires_grad=False)
x = torch.from_numpy(Xtr).type(dtype)
y = torch.from_numpy(ytr).type(dtype)
# x = torch.from_numpy(Xtr) # expected backend CPU and dtype Double



# w1 = Variable(torch.rand(1).type(dtype), requires_grad=True)
# w2 = Variable(torch.rand(1).type(dtype), requires_grad=True)
w1 = torch.rand(1, requires_grad=True) 
w2 = torch.rand(1, requires_grad=True)


learning_rate = 0.01

for epoch in range(1000):
    
    y_pred = x*w1 + w2
    
    loss = (y_pred - y).pow(2).mean()
    
    if epoch % 100 == 0:
        print(epoch, loss.item())
    
    loss.backward()
    
    w1.data -= learning_rate * w1.grad.data
    w2.data -= learning_rate * w2.grad.data

    w1.grad.data.zero_()
    w2.grad.data.zero_()
```
